```python
import time
import requests
from selenium import webdriver
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import pandas as pd
```

```python
driver = webdriver.Chrome(executable_path=r"F:/applications/chromedriver_win32/chromedriver.exe")

driver.get("https://medium.com/search?q=web%20development")
time.sleep(3)

scroll_pause_time = 3
screen_height = driver.execute_script("return window.screen.height;")

i = 1

while True:
    
    driver.execute_script("window.scrollTo(0, {screen_height}*{i});".format(screen_height=screen_height, i=i))
    
    i += 1
    
    time.sleep(scroll_pause_time)
    
    scroll_height = driver.execute_script("return document.body.scrollHeight;")
    
    if(screen_height)*i > scroll_height:
        break
        



soup=BeautifulSoup(driver.page_source,"html.parser")

links = soup.find_all("div",attrs={"class":"postArticle postArticle--short js-postArticle js-trackPostPresentation"})
print(len(links))



results = pd.DataFrame()

frame = []

for j in links:
    
    try:
        
        statement = j.find('div',attrs={"class":"section-content"}).find('h3').text.strip()
    
    except:
        
        statement = None
    
    
    
    try:
        tag = j.find('div',attrs={"class":"postMetaInline postMetaInline-authorLockup ui-captionStrong u-flex1 u-noWrapWithEllipsis"}).find('a',attrs={"class":"ds-link ds-link--styleSubtle link--darken link--accent u-accentColor--textNormal"}).text.strip()
        
    except:
        
        tag = None
           
           
    

     
           

    try:
        
        read_time = j.find('div',attrs={"class":"ui-caption u-fontSize12 u-baseColor--textNormal u-textColorNormal js-postMetaInlineSupplemental"}).find('span',attrs={"class":"readingTime"}).get('title')
    
    
    except:
        
        read_time = None
        
        
        
    try:
        
        date = j.find('div',attrs={"class":"ui-caption u-fontSize12 u-baseColor--textNormal u-textColorNormal js-postMetaInlineSupplemental"}).find("a", attrs={"class":"link link--darken"}).find('time').text[0:7].strip()
     
    except: 
        
        date = None
    
    
    
    try:
        
        year = j.find('div',attrs={"class":"ui-caption u-fontSize12 u-baseColor--textNormal u-textColorNormal js-postMetaInlineSupplemental"}).find("a", attrs={"class":"link link--darken"}).find("time").text[7:12].strip()
     
    except: 
        
        year = None
    
    
   
    try:
        
        claps = j.find('div', attrs={"class":"u-clearfix u-paddingTop10"}).find("span", attrs={"class":"u-relative u-background js-actionMultirecommendCount u-marginLeft5"}).find("button", attrs={"class":"button button--chromeless u-baseColor--buttonNormal js-multirecommendCountButton u-disablePointerEvents"}).text
        
    except:
        
        claps = None
        
    
    

        
        
    try:
        
        content = j.find('div', attrs={"class":"section-content"}).find('p', attrs={"class":"graf graf--p graf-after--h3 graf--trailing"}).text.strip()
        
        
    except:
        
        content = None
        
        
        
    try:
        
        comments = j.find('div', attrs={"class":"u-clearfix u-paddingTop10"}).find('div', attrs={"class":"buttonSet u-floatRight"}).find('a', attrs={"class":"button button--chromeless u-baseColor--buttonNormal"}).text
        
    except:
        
        comments = None
        
   
    
        
    temp_df = pd.DataFrame([[statement, content, tag, read_time, date, year, claps, comments]], columns=['Statement','Content','Publication','Read_Time',"Date","Year","Claps","comments"])
    
    results = results.append(temp_df)
    
    
results.to_csv('results_webdev.csv', index=False)

driver.close()    
  
 
```
